```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
anos = [2015,2016,2017,2018,2019]

df_qtde_mun = []
df_qtde_internet = []
df_qtde_agua_rede_publica = []
df_qtde_energia_rede_publica = []
df_qtde_esgoto_rede_publica = []
df_qtde_esgoto_fossa = []
df_qtde_almoxarifado = []
df_qtde_auditorio = []
df_qtde_banheiro_ei = []
df_qtde_banheiro_pne = []
df_qtde_biblioteca = []
df_qtde_lab_ciencias = []
df_qtde_lab_info = []
df_qtde_patio_coberto = []
df_qtde_parque_infantil = []
df_qtde_quadra_coberta = []
df_qtde_refeitorio = []
df_qtde_computador = []
df_qtde_despensa = []
df_qtde_creche = []
df_qtde_pre = []
df_qtde_ef_ai = []
df_qtde_ef_af = []
df_qtde_eja = []
```

```python
for ano in anos:
    df_esc = pd.read_csv(f'Dados/{ano}/ESCOLAS.CSV', sep = '|', encoding = 'latin1', decimal = ",")
    
    df_pmm_index = df_esc[(df_esc['CO_MUNICIPIO'] == 2307650) & (df_esc['TP_DEPENDENCIA'] == 3) & (df_esc['TP_SITUACAO_FUNCIONAMENTO'] == 1)]
    df_pmm = df_esc[(df_esc['CO_MUNICIPIO'] == 2307650) & (df_esc['TP_DEPENDENCIA'] == 3) & (df_esc['TP_SITUACAO_FUNCIONAMENTO'] == 1)]
        
    qtde_mun = df_pmm['CO_ENTIDADE'].count()
    qtde_internet = df_pmm['IN_INTERNET'].sum()
    qtde_agua_rede_publica = df_pmm['IN_AGUA_REDE_PUBLICA'].sum()
    qtde_energia_rede_publica = df_pmm['IN_ENERGIA_REDE_PUBLICA'].sum()
    qtde_esgoto_rede_publica = df_pmm['IN_ESGOTO_REDE_PUBLICA'].sum()
    qtde_esgoto_fossa = df_pmm['IN_ESGOTO_FOSSA'].sum()
    qtde_almoxarifado = df_pmm['IN_ALMOXARIFADO'].sum()
    qtde_auditorio = df_pmm['IN_AUDITORIO'].sum()
    qtde_banheiro_ei = df_pmm['IN_BANHEIRO_EI'].sum()
    qtde_banheiro_pne = df_pmm['IN_BANHEIRO_PNE'].sum()
    qtde_biblioteca = df_pmm['IN_BIBLIOTECA'].sum()
    qtde_lab_ciencias = df_pmm['IN_LABORATORIO_CIENCIAS'].sum()
    qtde_lab_info = df_pmm['IN_LABORATORIO_INFORMATICA'].sum()
    qtde_patio_coberto = df_pmm['IN_PATIO_COBERTO'].sum()   
    qtde_parque_infantil = df_pmm['IN_PARQUE_INFANTIL'].sum()
    qtde_quadra_coberta = df_pmm['IN_QUADRA_ESPORTES_COBERTA'].sum()  
    qtde_refeitorio = df_pmm['IN_REFEITORIO'].sum()
    qtde_computador = df_pmm['IN_COMPUTADOR'].sum()
    qtde_despensa = df_pmm['IN_DESPENSA'].sum()
    
    # Categorias
    qtde_creche = df_pmm['IN_COMUM_CRECHE'].sum()
    qtde_pre = df_pmm['IN_COMUM_PRE'].sum()
    qtde_ef_af = df_pmm['IN_COMUM_FUND_AF'].sum()
    qtde_ef_ai = df_pmm['IN_COMUM_FUND_AI'].sum()
    qtde_eja = df_pmm['IN_EJA'].sum()        
      
    # Atributos de Infraestrutura
    df_qtde_mun.append(qtde_mun)
    df_qtde_internet.append(qtde_internet)
    df_qtde_agua_rede_publica.append(qtde_agua_rede_publica)
    df_qtde_energia_rede_publica.append(qtde_energia_rede_publica)
    df_qtde_esgoto_rede_publica.append(qtde_esgoto_rede_publica)
    df_qtde_esgoto_fossa.append(qtde_esgoto_fossa)
    df_qtde_almoxarifado.append(qtde_almoxarifado)
    df_qtde_auditorio.append(qtde_auditorio)
    df_qtde_banheiro_ei.append(qtde_banheiro_ei)
    df_qtde_banheiro_pne.append(qtde_banheiro_pne)
    df_qtde_biblioteca.append(qtde_biblioteca)
    df_qtde_lab_ciencias.append(qtde_lab_ciencias)
    df_qtde_lab_info.append(qtde_lab_info)
    df_qtde_patio_coberto.append(qtde_patio_coberto)
    df_qtde_parque_infantil.append(qtde_parque_infantil)
    df_qtde_quadra_coberta.append(qtde_quadra_coberta)
    df_qtde_refeitorio.append(qtde_refeitorio)
    df_qtde_computador.append(qtde_computador)
    df_qtde_despensa.append(qtde_despensa)
    
    df_qtde_creche.append(qtde_creche)
    df_qtde_pre.append(qtde_pre)
    df_qtde_ef_ai.append(qtde_ef_ai)
    df_qtde_ef_af.append(qtde_ef_af)
    df_qtde_eja.append(qtde_eja)
```

```python
#  infraestrutura

df = {}
df['Total'] = df_qtde_mun
df['Internet'] = df_qtde_internet
df['Agua'] = df_qtde_agua_rede_publica
df['Energia'] = df_qtde_energia_rede_publica
df['Esgoto_Rede'] = df_qtde_esgoto_rede_publica
df['Esgoto_Fossa'] = df_qtde_esgoto_fossa
df['Almoxarifado'] = df_qtde_almoxarifado
df['Auditorio'] = df_qtde_auditorio
df['Banheiro_EI'] = df_qtde_banheiro_ei
df['Banheiro_PNE'] = df_qtde_banheiro_pne
df['Biblioteca'] = df_qtde_biblioteca
df['Laboratorio_Ciencias'] = df_qtde_lab_ciencias
df['Laboratorio_Informatica'] = df_qtde_lab_info
df['Patio_Coberto'] = df_qtde_patio_coberto
df['Parque_Infantil'] = df_qtde_parque_infantil
df['Quadra_Coberta'] = df_qtde_quadra_coberta
df['Refeitorio'] = df_qtde_refeitorio
df['Computador'] = df_qtde_computador
df['Despensa'] = df_qtde_despensa

df_infra = pd.DataFrame (data = df, index = [anos])
print('Mapeamento de Infraestrutura das Escolas Municicipais de Maracanaú no período de 2015-2019')
df_infra.to_csv('mapeamento_infra_2015_2019.csv')
df_infra
```

```python
df_pmm_space = df_pmm_space[['IN_ESGOTO_REDE_PUBLICA',
                             'IN_AUDITORIO',
                             'IN_BIBLIOTECA',
                             'IN_LABORATORIO_CIENCIAS',
                             'IN_LABORATORIO_INFORMATICA',
                             'QT_SALAS_UTILIZADAS',
                             'IN_EQUIP_LOUSA_DIGITAL',
                             'IN_DESKTOP_ALUNO',
                             'IN_COMP_PORTATIL_ALUNO',
                             'IDE_Medio_EF']]
df_pmm_space
```

```python
df_pmm_space = df_pmm_space.fillna(value=0)
```
