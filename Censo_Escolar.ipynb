```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
anos = [2015,2016,2017,2018,2019]

cod_dependencia = 3 # Municipal
cod_situacao = 1 #Em atividade

# dicionário dos municípios

municipios = [[2304400,'FORTALEZA'],
              [2303709,'CAUCAIA'],
              [2304285,'EUSEBIO'],
              [2307700,'MARANGUAPE'],
              [2301000,'AQUIRAZ'],
              [2309706,'PACATUBA'],
              [2303501,'CASCAVEL'],
              [2312403,'SGAMARANTE'], 
              [2312908,'SOBRAL'],
              [2307650,'MARACANAU']]
```

```python
lt_ano = []
lt_municipio = []
lt_qtde_mun = []
lt_qtde_internet = []
lt_qtde_agua_rede_publica = []
lt_qtde_energia_rede_publica = []
lt_qtde_esgoto_rede_publica = []
lt_qtde_esgoto_fossa = []
lt_qtde_almoxarifado = []
lt_qtde_auditorio = []
lt_qtde_banheiro_ei = []
lt_qtde_banheiro_pne = []
lt_qtde_biblioteca = []
lt_qtde_lab_ciencias = []
lt_qtde_lab_info = []
lt_qtde_patio_coberto = []
lt_qtde_parque_infantil = []
lt_qtde_quadra_coberta = []
lt_qtde_refeitorio = []
lt_qtde_computador = []
lt_qtde_despensa = []
lt_qtde_creche = []
lt_qtde_pre = []
lt_qtde_ef_ai = []
lt_qtde_ef_af = []
lt_qtde_eja = []

for ano in anos:
    
    df_esc = pd.read_csv(f'Dados/{ano}/ESCOLAS.CSV', sep = '|', encoding = 'latin1', decimal = ",")
    
    for cod in range(0,len(municipios)):
               
        df_pmm_index = df_esc[(df_esc['CO_MUNICIPIO'] == municipios[cod][0]) & (df_esc['TP_DEPENDENCIA'] == cod_dependencia) & (df_esc['TP_SITUACAO_FUNCIONAMENTO'] == cod_situacao)]
        df_pmm = df_esc[(df_esc['CO_MUNICIPIO'] == municipios[cod][0]) & (df_esc['TP_DEPENDENCIA'] == cod_dependencia) & (df_esc['TP_SITUACAO_FUNCIONAMENTO'] == cod_situacao)]

        qtde_mun = df_pmm['CO_ENTIDADE'].count()
        qtde_internet = df_pmm['IN_INTERNET'].sum()
        qtde_agua_rede_publica = df_pmm['IN_AGUA_REDE_PUBLICA'].sum()
        qtde_energia_rede_publica = df_pmm['IN_ENERGIA_REDE_PUBLICA'].sum()
        qtde_esgoto_rede_publica = df_pmm['IN_ESGOTO_REDE_PUBLICA'].sum()
        qtde_esgoto_fossa = df_pmm['IN_ESGOTO_FOSSA'].sum()
        qtde_almoxarifado = df_pmm['IN_ALMOXARIFADO'].sum()
        qtde_auditorio = df_pmm['IN_AUDITORIO'].sum()
        qtde_banheiro_ei = df_pmm['IN_BANHEIRO_EI'].sum()
        qtde_banheiro_pne = df_pmm['IN_BANHEIRO_PNE'].sum()
        qtde_biblioteca = df_pmm['IN_BIBLIOTECA'].sum()
        qtde_lab_ciencias = df_pmm['IN_LABORATORIO_CIENCIAS'].sum()
        qtde_lab_info = df_pmm['IN_LABORATORIO_INFORMATICA'].sum()
        qtde_patio_coberto = df_pmm['IN_PATIO_COBERTO'].sum()   
        qtde_parque_infantil = df_pmm['IN_PARQUE_INFANTIL'].sum()
        qtde_quadra_coberta = df_pmm['IN_QUADRA_ESPORTES_COBERTA'].sum()  
        qtde_refeitorio = df_pmm['IN_REFEITORIO'].sum()
        qtde_computador = df_pmm['IN_COMPUTADOR'].sum()
        qtde_despensa = df_pmm['IN_DESPENSA'].sum()
    
        # Categorias
        qtde_creche = df_pmm['IN_COMUM_CRECHE'].sum()
        qtde_pre = df_pmm['IN_COMUM_PRE'].sum()
        qtde_ef_af = df_pmm['IN_COMUM_FUND_AF'].sum()
        qtde_ef_ai = df_pmm['IN_COMUM_FUND_AI'].sum()
        qtde_eja = df_pmm['IN_EJA'].sum()        

        # Atributos de Infraestrutura

        lt_ano.append(ano)
        lt_municipio.append(municipios[cod][1])
        lt_qtde_mun.append(qtde_mun)
        lt_qtde_internet.append(qtde_internet)
        lt_qtde_agua_rede_publica.append(qtde_agua_rede_publica)
        lt_qtde_energia_rede_publica.append(qtde_energia_rede_publica)
        lt_qtde_esgoto_rede_publica.append(qtde_esgoto_rede_publica)
        lt_qtde_esgoto_fossa.append(qtde_esgoto_fossa)
        lt_qtde_almoxarifado.append(qtde_almoxarifado)
        lt_qtde_auditorio.append(qtde_auditorio)
        lt_qtde_banheiro_ei.append(qtde_banheiro_ei)
        lt_qtde_banheiro_pne.append(qtde_banheiro_pne)
        lt_qtde_biblioteca.append(qtde_biblioteca)
        lt_qtde_lab_ciencias.append(qtde_lab_ciencias)
        lt_qtde_lab_info.append(qtde_lab_info)
        lt_qtde_patio_coberto.append(qtde_patio_coberto)
        lt_qtde_parque_infantil.append(qtde_parque_infantil)
        lt_qtde_quadra_coberta.append(qtde_quadra_coberta)
        lt_qtde_refeitorio.append(qtde_refeitorio)
        lt_qtde_computador.append(qtde_computador)
        lt_qtde_despensa.append(qtde_despensa)

        lt_qtde_creche.append(qtde_creche)
        lt_qtde_pre.append(qtde_pre)
        lt_qtde_ef_ai.append(qtde_ef_ai)
        lt_qtde_ef_af.append(qtde_ef_af)
        lt_qtde_eja.append(qtde_eja)
        
```

```python
#  infraestrutura

dc = {}
dc['Ano'] = lt_ano
dc['Municipio'] = lt_municipio
dc['Total'] = lt_qtde_mun
dc['Internet'] = lt_qtde_internet
dc['Agua'] = lt_qtde_agua_rede_publica
dc['Energia'] = lt_qtde_energia_rede_publica
dc['Esgoto_Rede'] = lt_qtde_esgoto_rede_publica
dc['Esgoto_Fossa'] = lt_qtde_esgoto_fossa
dc['Almoxarifado'] = lt_qtde_almoxarifado
dc['Auditorio'] = lt_qtde_auditorio
dc['Banheiro_EI'] = lt_qtde_banheiro_ei
dc['Banheiro_PNE'] = lt_qtde_banheiro_pne
dc['Biblioteca'] = lt_qtde_biblioteca
dc['Laboratorio_Ciencias'] = lt_qtde_lab_ciencias
dc['Laboratorio_Informatica'] = lt_qtde_lab_info
dc['Patio_Coberto'] = lt_qtde_patio_coberto
dc['Parque_Infantil'] = lt_qtde_parque_infantil
dc['Quadra_Coberta'] = lt_qtde_quadra_coberta
dc['Refeitorio'] = lt_qtde_refeitorio
dc['Computador'] = lt_qtde_computador
dc['Despensa'] = lt_qtde_despensa

#df_infra = pd.DataFrame (data = df, index = [anos])
df_infra = pd.DataFrame (data = dc)
print('Mapeamento de Infraestrutura das Escolas Municicipais no período de 2015-2019')
df_infra.to_csv('mapeamento_infra_2015_2019.csv')
df_infra.head()
```
